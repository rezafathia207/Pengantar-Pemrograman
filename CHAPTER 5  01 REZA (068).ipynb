```python
import pandas as pd  # Untuk membaca dan memproses dataset Excel
from scipy.stats.mstats import gmean, hmean  # Untuk menghitung geometric & harmonic mean
import matplotlib.pyplot as plt  # Untuk plotting grafik

# 1️⃣ Baca dataset
def read_dataset(filename, sheet_name):
    return pd.read_excel(filename, sheet_name=sheet_name)  # Membaca sheet Excel dan mengembalikan DataFrame

# 2️⃣ Hitung mean
def arithmetic_mean(data):
    return data.mean()  # Rata-rata aritmetika

def geometric_mean(data):
    return gmean(data)  # Rata-rata geometrik

def harmonic_mean(data):
    return hmean(data)  # Rata-rata harmonik

# 3️⃣ Print mean
def print_mean(name, value):
    print('-------')  # Separator
    print(name)  # Nama mean
    print("{0:.1f} [ppm]".format(value))  # Nilai mean dengan format 1 desimal

# 4️⃣ Plot histogram
def create_histogram(data, bins='auto', density=True, edgecolor='k', alpha=0.8):
    fig, ax = plt.subplots(figsize=(6,4))  # Membuat figure dan axis
    ax.hist(data, bins=bins, density=density, edgecolor=edgecolor, alpha=alpha, label='Measurements Hist')  # Plot histogram
    return fig, ax  # Kembalikan figure & axis untuk modifikasi selanjutnya

# 5️⃣ Tambah garis mean ke histogram
def add_mean_line(ax, value, color, label):
    ax.axvline(value, color=color, label=label, linewidth=2)  # Tambahkan garis vertikal untuk mean

# 6️⃣ Tambah label, legend, dan tampilkan plot
def finalize_plot(ax, xlabel, ylabel, title=''):
    ax.set_xlabel(xlabel)  # Label sumbu x
    ax.set_ylabel(ylabel)  # Label sumbu y
    ax.set_title(title)  # Judul plot
    ax.legend()  # Tampilkan legend
    ax.grid(True)  # Tampilkan grid
    plt.show()  # Tampilkan plot

# === MAIN ===
my_dataset = read_dataset('Smith_glass_post_NYT_data.xlsx', 'Supp_traces')  # Baca dataset
data = my_dataset['Zr']  # Ambil kolom Zr

# Hitung semua mean
a_mean = arithmetic_mean(data)  # Arithmetic mean
g_mean = geometric_mean(data)  # Geometric mean
h_mean = harmonic_mean(data)  # Harmonic mean

# Print semua mean
print_mean('Arithmetic mean', a_mean)  # Tampilkan arithmetic mean
print_mean('Geometric mean', g_mean)  # Tampilkan geometric mean
print_mean('Harmonic mean', h_mean)  # Tampilkan harmonic mean

# Buat histogram
fig, ax = create_histogram(data)  # Buat histogram

# Tambahkan semua garis mean
add_mean_line(ax, a_mean, 'purple', 'Arithmetic mean')  # Tambahkan garis arithmetic mean
add_mean_line(ax, g_mean, 'orange', 'Geometric mean')  # Tambahkan garis geometric mean
add_mean_line(ax, h_mean, 'green', 'Harmonic mean')  # Tambahkan garis harmonic mean

# Finalisasi plot
finalize_plot(ax, 'Zr [ppm]', 'Probability density', 'Histogram of Zr Measurements')  # Label, legend, grid, tampilkan plot
```
